```python
threshold = 99

missing_percent = (df.isnull().sum() / len(df)) * 100

cols_to_drop = missing_percent[missing_percent >= threshold].index

df_cleaned = df.drop(columns = cols_to_drop)
```

```python
df_cleaned.info()
df_cleaned 
```

```python
not_null_info = pd.DataFrame({
    'not_null_count': df_cleaned.notnull().sum(),
    'not_null_percent': (df_cleaned.notnull().sum() / len(df_cleaned)) * 100,
    'data_type': df_cleaned.dtypes
}).sort_values(by='not_null_percent', ascending=False)

not_null_info
```
